```python
import plotly.express as px

fig = px.box(df, x='Winner Pts', orientation='h', title='Winner Pts')

fig.show()
```

The above Histogram shows the range in Winner Points. It shows a median of 30.5, with the minimum and maximum being 13 and 55. This gives us a range of 42.

```python
fig = px.histogram(df, x='Loser Pts', title='Loser Points Distribution')
fig.show()
```

```python
fig = px.density_heatmap(
    df,
    x='Winner',
    y='Winner Pts',
    title='Winner vs Winner Pts',
    height=600
)
fig.show()
```

```python
fig = px.density_heatmap(
    df,
    x='Loser',
    y='Loser Pts',
    title='Loser vs Loser Pts',
    height=600
)
fig.show()
```

```python
fig = px.line(df, x='Date', y='Winner Pts', title='Winner Points per year')
fig.show()
```

```python
fig = px.scatter(
    df,
    x='Date',
    y='Winner Pts',
    title='Winner Points Per Year',
    trendline='ols'
)
fig.show()
```

```python
teams = df['Winner'].append(df['Loser'], ignore_index=True)
teams.to_frame().drop_duplicates().rename(columns={
    0: 'Team'
})
```

Above I have visualized the data using a line and a scatter graph, with the graphs showing the Winners Points per year. The graphs show that there is no clear correlation between the Date of the superbowl and the Winners Points. From this I can answer one of my questions, there is no trend between the Date and the Winners Points.

```python
df_winners_sum = df['Winner'].value_counts().to_frame().reset_index() \
  .rename(columns={
      'index': 'Team',
      'Winner': 'Wins'
  })

df_winners_sum
```

```python
fig = px.pie(
    df_winners_sum,
    names='Team',
    values='Wins',
    title='Winners breakdown',
)

fig.show()
```
